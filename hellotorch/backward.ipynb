```python
import torch

# 1. 创建自变量 x
# 关键点 1: 必须是浮点数 (3.0)，整数不能求导
# 关键点 2: requires_grad=True 告诉 PyTorch："嘿，帮我盯着这个变量，我待会儿要求它的导数！"
x = torch.tensor(3.0, requires_grad=True)

print(f"初始值 x = {x}")
print(f"x 是否开启追踪? {x.requires_grad}")

```

```python
# 2. 定义计算图 (也就是函数 y = x^2)
# PyTorch 会在后台记录下这个计算过程：y 是怎么从 x 变过来的
y = x ** 2

print(f"计算结果 y = x^2 = {y}") 
# 你会看到 y=9，而且它也有 grad_fn 属性，说明它是计算出来的

# 3. 核心魔法：反向传播 (Autograd)
# 这一步就相当于 PyTorch 在脑子里飞快地算了一遍微积分：dy/dx = 2x
y.backward()
help(y.backward)
# 4. 查看导数 (梯度)
# 导数就存在 x.grad 里面！
print(f"\n🎉 自动计算的导数 dy/dx (在 x=3 处) = {x.grad}")
```
